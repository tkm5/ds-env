```python
df_data = pd.DataFrame(bc.data, columns=bc.feature_names)
df_tgt = pd.DataFrame(bc.target, columns=['class'])
df = pd.concat([df_data, df_tgt], axis=1)
```

```python
df_pu = df.loc[:, ['worst perimeter', 'worst concave points', 'worst radius', 'mean concave points', 'class']]
sns.pairplot(data=df_pu, hue='class', diag_kind='kde', markers='+',plot_kws={'alpha':0.2})
```
