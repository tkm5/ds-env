# matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5]
# y = [1 ,4, 9, 16, 25]
y = np.array(x) ** 2
plt.plot(x, y)
```

```python
y
```

```python
x = np.linspace(-10, 10, 100)
z = 1 / (1 + np.exp(-x))
plt.plot(x, z, 'b--')
```

```python
df = pd.DataFrame({'a':[1, 2, 3, 4, 5, 6],
                   'b':[10, 15, 40, 100, 200, 400]})
```

```python
plt.plot('a' , 'b' , data=df)
```

```python
plt.plot(x, z,
        color='red',
        lw=3,
        ls='--',
        marker='o',
        markersize=2,
        markerfacecolor='blue',
        markeredgecolor='green',
        markeredgewidth=4,
        alpha=0.2)
```

```python
plt.plot(x, z, label='sigmoid')
plt.plot(x, z**2, label='sigmoid*2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=0)
plt.xticks([1, 2, 3], rotation=20)
plt.yticks([0, 0.5, 1], rotation=20)
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df = df[~df['release_date'].isna()]
```

```python
df['year'] = df['release_date'].apply(lambda x: int(x[:4]))
```

```python
plt.plot(df.groupby('year').count()['id'])
plt.xlabel('year')
plt.ylabel('count')
```

```python
df = df[df['revenue'] != 0]
```

```python
plt.plot(df.groupby('year').median()['revenue'], label='median')
plt.plot(df.groupby('year').mean()['revenue'], label='mean')
plt.plot(df.groupby('year')['revenue'].max(), label='max')
plt.legend()
```

```python
x = np.linspace(-10, 10, 100)
plt.subplot(1, 2, 1)
plt.plot(x, x**2)

plt.subplot(1, 2, 2)
plt.plot(x, x**3)
```

```python
plt.plot(x, x**2)
```

```python
plt.plot(x, x**3)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x, x**2)
ax2.plot(x, x**3)
```

```python
fig, axes = plt.subplots(2, 2)
axes[0, 0].plot(x, x**2, label='x**2')
axes[0, 0].set_xlabel('x level1')
axes[0, 0].legend()
axes[0, 1].plot(x, x**7)
axes[1, 0].plot(x, x**8)
axes[1, 1].plot(x, x**9)
fig.tight_layout()
```

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
plt.hist(df['vote_average'], bins=50)
```

```python
fig, axes = plt.subplots()
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
axes.hist(df['profit'])
```
