## split num_cal and cat_cal

```python
# create numerical data frame
numerical_cols = [cname for cname in train_df.columns
                  if train_df[cname].dtype == 'float64']
```

```python
train_df_numcols = train_df[numerical_cols].copy()
train_df_numcols
```

```python
# create categorical data frame
categorical_cols = [cname for cname in train_df.columns
                    if train_df[cname].dtype == "object"]
```

```python
train_df_catcols = train_df[categorical_cols].copy()
train_df_catcols
```
